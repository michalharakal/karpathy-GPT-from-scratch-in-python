```python
import torch
import matplotlib.pyplot as plt

# Create a tensor of size 100
tensor = torch.empty(100)

# Initialize the tensor with values drawn from a normal distribution
# with mean=0 and standard deviation=1
torch.nn.init.normal_(tensor, mean=0, std=1)

# Convert tensor to numpy array for plotting
values = tensor.numpy()

# Plotting
plt.hist(values, bins=10, alpha=0.7, label='Histogram of Initialized Values')
plt.title('Distribution of 100 Normally Initialized Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
